Load dataset

```python
import pandas as pd
import numpy as np
df = pd.read_csv('/content/drive/MyDrive/ML Dataset/titanic_train.csv')
df.head()
```

**Drop unwanted data**

```python
df.drop("Cabin", axis=1, inplace=True)
df.drop("Embarked", axis=1, inplace=True)
```

```python
updated_df = df
updated_df['Age']=updated_df['Age'].fillna(updated_df['Age'].mean())
updated_df.info()
```

```python
updated_df = df.dropna(axis=1)
updated_df.info()
```

Delcare update_df through survival

```python
updated_df = df.dropna(axis=0)
y1 = updated_df['Survived']
updated_df.drop("Survived",axis=1,inplace=True)
```

**Import libraries**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

**Distribution plot of histogram**

```python
sns.histplot(updated_df['Age'], bins=10, kde=True)
```

```python
sns.catplot(x="Categorical_Age", y="Age", kind="box", data=updated_df)
```

**Line Plot of time**

```python
test_df = pd.DataFrame(dict(time=pd.date_range("2017-1-1", periods=500),
value=np.random.randn(500).cumsum()))
sns.relplot(x="time", y="value", kind="line", data=test_df ,height=4,
aspect=15/6)
```
